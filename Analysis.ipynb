```python
# -------------------------------
# STEP 3: Libraries import
# -------------------------------
import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------
# STEP 4: Dataset load
# -------------------------------
data = pd.read_csv(
    "https://raw.githubusercontent.com/GauravDataAI/COVID-19-Data-Analysis/main/Covid_Data.csv.csv"
)

# -------------------------------
# STEP 5: Dataset output dikhao
# -------------------------------
print("\n--- Full Dataset ---")
print(data)

# -------------------------------
# STEP 6: Global summary
# -------------------------------
print("\n--- Global COVID Summary ---")
print("Total Confirmed:", data["Confirmed"].sum())
print("Total Recovered:", data["Recovered"].sum())
print("Total Deaths:", data["Deaths"].sum())

# -------------------------------
# STEP 7: Recovery & Death Rates
# -------------------------------
data["Recovery_Rate (%)"] = (data["Recovered"] / data["Confirmed"]) * 100
data["Death_Rate (%)"] = (data["Deaths"] / data["Confirmed"]) * 100

print("\n--- Country-wise Rates ---")
print(data[["Country", "Recovery_Rate (%)", "Death_Rate (%)"]])

# -------------------------------
# STEP 8: Graphs
# -------------------------------

# 1 Confirmed cases
plt.figure(figsize=(10,5))
plt.bar(data["Country"], data["Confirmed"], color='orange')
plt.title("Country-wise Confirmed Cases")
plt.xticks(rotation=45)
plt.show()

# 2 Recovery vs Death
plt.figure(figsize=(10,5))
plt.bar(data["Country"], data["Recovered"], label="Recovered", color='green')
plt.bar(data["Country"], data["Deaths"], bottom=data["Recovered"], label="Deaths", color='red')
plt.xticks(rotation=45)
plt.legend()
plt.title("Recovery vs Death")
plt.show()

# 3 Death pie chart
plt.figure(figsize=(7,7))
plt.pie(data["Deaths"], labels=data["Country"], autopct="%1.1f%%", startangle=140)
plt.title("Death Distribution")
plt.show()

# -------------------------------
# STEP 9: Key Observations
# -------------------------------
print("\n--- Key Observations ---")
print("USA has highest confirmed cases.")
print("India shows strong recovery.")
print("Deaths are lower compared to recoveries.")
```
